```python
# 인공지능의 이해: 과제1

# 문제1.층을 하나 늘려 예측 능력이 어떻게 변화하는지 확인하고 최종 손실값을 작성하시오.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로우와 케라스를 사용하자
import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential( [
    keras.layers.Dense(6, activation= 'sigmoid'),
    keras.layers.Dense(4, activation= 'sigmoid'),
    keras.layers.Dense(1, activation= 'sigmoid'),
    # 층을 하나 늘려 예측 능력의 변화를 살펴보자
    keras.layers.Dense(1, activation= 'sigmoid'),
])

optimizer = keras.optimizers.SGD(learning_rate=5.0)
model.compile(optimizer=optimizer, loss='mse')

data_loc = 'https://github.com/dknife/ML/raw/main/data/'

df = pd.read_csv(data_loc+'nonlinear.csv')
X = df['x'].to_numpy()
y_label = df['y'].to_numpy()

model.fit(tf.expand_dims(X, axis=-1), y_label, epochs=100)

# 입력을 2차원 벡터로 변형
domain = np.linspace(0, 1, 100).reshape(-1,1)
y_hat = model.predict(domain)
plt.scatter(df['x'], df['y'])

# 최종 손실값은 0.3981 이다.
plt.scatter(domain, y_hat, color='r')
```

```python
# 문제2.활성화함수는 시그모이드뿐만 아니라 여러 종류의 함수를 사용할수 있다.
# 각 계층의 활성화 함수를 tanh를 쓰고 학습률을 0.1로 바꾸어 어떻게 될지 살펴 보자.
# 그리고, 최종 손실값을 작성하시오.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로우와 케라스를 사용하자
import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential( [
    # 활성화 함수 tanh를 사용하자
    keras.layers.Dense(4, activation= 'tanh'),
    keras.layers.Dense(4, activation= 'tanh'),
    keras.layers.Dense(4, activation= 'tanh'),
    keras.layers.Dense(1, activation= 'tanh'),
])

# 학습률은 0.1으로 바꾸자
optimizer = keras.optimizers.SGD(learning_rate=0.1)
model.compile(optimizer=optimizer, loss='mse')

data_loc = 'https://github.com/dknife/ML/raw/main/data/'

df = pd.read_csv(data_loc+'nonlinear.csv')
X = df['x'].to_numpy()
y_label = df['y'].to_numpy()

model.fit(tf.expand_dims(X, axis=-1), y_label, epochs=100)

domain = np.linspace(0, 1, 100).reshape(-1,1) # 입력은 2차원 벡터로 변형
y_hat = model.predict(domain)
plt.scatter(df['x'], df['y'])

# 최종 손실값은 0.2978 이다.
plt.scatter(domain, y_hat, color='r')
```

```python
# 문제3.앞의 문제를 풀면서 사용한 학습률을 0.1, 1.0, 2.0, 5.0으로 변경하여
# 학습을 수행한 결과에서 각각 최종 손실값을 작성하시오.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 텐서플로우와 케라스를 사용
import tensorflow as tf
from tensorflow import keras

model = keras.models.Sequential( [
    keras.layers.Dense(4, activation= 'tanh'),
    keras.layers.Dense(4, activation= 'tanh'),
    keras.layers.Dense(4, activation= 'tanh'),
    keras.layers.Dense(1, activation= 'tanh'),
])

# 학습률을 0.1, 1.0, 2.0, 5.0으로 변경하자
# optimizer = keras.optimizers.SGD(learning_rate=0.1)
# optimizer = keras.optimizers.SGD(learning_rate=1.0)
# optimizer = keras.optimizers.SGD(learning_rate=2.0)
optimizer = keras.optimizers.SGD(learning_rate=5.0)
model.compile(optimizer=optimizer, loss='mse')

data_loc = 'https://github.com/dknife/ML/raw/main/data/'

df = pd.read_csv(data_loc+'nonlinear.csv')
X = df['x'].to_numpy()
y_label = df['y'].to_numpy()

model.fit(tf.expand_dims(X, axis=-1), y_label, epochs=100)

domain = np.linspace(0, 1, 100).reshape(-1,1) # 입력은 2차원 벡터로 변형
y_hat = model.predict(domain)
plt.scatter(df['x'], df['y'])

# 학습률 = 0.1 -> 최종 손실값 = 0.2796
# 학습률 = 1.0 -> 최종 손실값 = 0.3438
# 학습률 = 2.0 -> 최종 손실값 = 1.2036
# 학습률 = 5.0 -> 최종 손실값 = 2.1390
plt.scatter(domain, y_hat, color='r')
```
